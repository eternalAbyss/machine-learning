# Introduction to Matplotlib

```python
# Importing the Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = [1,2,3,4]
y = [11,22,33,44]

# Plotting a basic figure
plt.plot(x,y)
```

```python
# 1st method
fig = plt.figure() # Creates a figure
ax = fig.add_subplot() # adds some axes
plt.show()
```

```python
# 2nd method
fig = plt.figure() # Creates a figure
ax = fig.add_axes([1,1,1,1])
ax.plot(x, y) # Add some data
plt.show()
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

## Matplotlib example workflow

```python
# Preparing the Data
x = [1,2,3,4]
y = [10,21,36,48]

# Setup plot
fig,ax = plt.subplots(figsize=(10,10))

# Plot data
ax.plot(x, y)

# Customize plot
ax.set(title="Simple Plot",
       xlabel="X-axis",
       ylabel="Y-axis")

# Save and Show (You save the whole figure)
fig.savefig("./sample-plot.png")
```

## Making figures with NumPy Arrays
We will be making:
* Line Plot
* Scatter Plot
* Bar Plot
* Histogram
* Subplot

```python
# Create some data and create a line plot 
x = np.linspace(0,10,100)
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use the same data and create a scatter plot
fig, ax = plt.subplots()
ax.scatter( x, np.exp(x))
plt.show()
```

```python
# Another scatter plot 
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
plt.show()
```

```python
# Make a plot from dictionary
nut_butter_prices = {
    'Peanut butter': 10,
    'Almond butter': 22,
    'Cashew butter': 30}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(
    title="Dan's nut butter prices", 
    xlabel="Type of butter",
    ylabel="Cost of 1kg jar($)")
plt.show()
```

```python
# Horizontal Bar graph
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values())) # For horizontal bar graphs we need lists
ax.set(
    title="Dan's nut butter prices",
    xlabel="Cost per kg ($)",
    ylabel="Type of butter"
)
plt.show()
```

```python
# Make some data for Histogram and plot it
x = np.random.randint(0,1000,size=(20))
fig, ax = plt.subplots()
ax.hist(x)
ax.set(
    title="Energy Consumption",
    xlabel="Watt",
    ylabel="Hours"
    )
plt.show()
```

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(20,10))

# Plot to each different axis
ax1.plot(nut_butter_prices.keys(), nut_butter_prices.values())
ax2.hist(nut_butter_prices.values())
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
plt.show()
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(20,10)
                      )

# Plotting to each different index
ax[0,0].plot(x)
ax[0,1].scatter(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(nut_butter_prices.values())
plt.show()
```

### Which should we use pyplot or matplotlib OO method?

* When plotting something quickly, okay to use pyplot method
* When plotting something more advanced, use the OO method

```python
over_50 = heart_desease[heart_desease['age']>50]
len(over_50)
```

```python
# Pyplot method 
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
# OO method mixed with pyplot method

fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax)
# ax.set_xlim([45,100])
plt.show()
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data 
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

# Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel='Age',
       ylabel='Cholestrol',
       )

# Add a legend
ax.legend(*scatter.legend_elements(), title="target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), 
           linestyle='--')
plt.show()
```

## Creating a subplot of Age, cholestrol and Maximum heart rate

```python
# intializing
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)
# Plotting the first scatter plot
scatter1 = ax0.scatter(
         x=over_50['age'],
         y=over_50['chol'],
         c=over_50['target'])

# Customizing the first plot
ax0.set(
        ylabel='Cholestrol',
        title='Scatter plot of Age Vs Cholestrol')

# Adding legends to the first plot
ax0.legend(*scatter.legend_elements(), title='target')

# Adding a line portraying the mean of cholestrol
ax0.axhline(over_50['chol'].mean(),
            linestyle='--')

# Plotting the second plot
scatter2 = ax1.scatter(
         x=over_50['age'],
         y=over_50['thalach'],
         c=over_50['target'])

# Customizing the second plot
ax1.set(xlabel='Age',
        ylabel='Maximum Heart rate',
        title='Scatter plot of Age Vs Maximum Heart Rate')

# Adding legends to the second plot
ax1.legend(*scatter.legend_elements(), title='target')
ax1.axhline(over_50['thalach'].mean(),
            linestyle="--")

# Set the title of the figure 
fig.suptitle("Heart Disease Analysis", fontsize=16)

plt.show()
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data 
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter') # This changes the color in the plot

# Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel='Age',
       ylabel='Cholestrol',
       )

# Add a legend
ax.legend(*scatter.legend_elements(), title="target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), 
           linestyle='--')
plt.show()
```

```python
# Customizing the x and y axis limits

# intializing
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)
# Plotting the first scatter plot
scatter1 = ax0.scatter(
         x=over_50['age'],
         y=over_50['chol'],
         c=over_50['target'],
         cmap='winter')

# Customizing the first plot
ax0.set(
        ylabel='Cholestrol',
        title='Scatter plot of Age Vs Cholestrol',
        xlim=[50,80],
        ylim=[100,600])

# Adding legends to the first plot
ax0.legend(*scatter.legend_elements(), title='target')

# Adding a line portraying the mean of cholestrol
ax0.axhline(over_50['chol'].mean(),
            linestyle='--')

# Cleaning the first plot


# Plotting the second plot
scatter2 = ax1.scatter(
         x=over_50['age'],
         y=over_50['thalach'],
         c=over_50['target'],
         cmap='winter')

# Customizing the second plot
ax1.set(xlabel='Age',
        ylabel='Maximum Heart rate',
        title='Scatter plot of Age Vs Maximum Heart Rate',
        xlim=[50,80],
        ylim=[60,200])

# Adding legends to the second plot
ax1.legend(*scatter.legend_elements(), title='target')
ax1.axhline(over_50['thalach'].mean(),
            linestyle="--")

# Set the title of the figure 
fig.suptitle("Heart Disease Analysis", fontsize=16)

plt.show()
```
